## Perform histogram equalization of a gray scale image.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('Week2_Q3_image.tif', 0)

#display the image that is loaded
fig = plt.figure(figsize=(3, 3))
plt.imshow(image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

## Visualizing Image Histogram with Matplotlib

 - This code computes and displays the histogram of an input image, showing the distribution of pixel intensities as a bar chart, which is a valuable tool in computer vision for analyzing image characteristics like brightness and contrast.

```python
histogram,bins = np.histogram(image.flatten(), bins=256, range=[0, 256])
#visualize the histogram in a bar chart

plt.figure(figsize=(6, 4))
plt.bar(bins[:-1], histogram, width=1, color='gray')
plt.xlabel('Intensity Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.xlim([0, 257])

plt.tight_layout()
plt.show()
```

```python
equalized_image = cv2.equalizeHist(image)
```

```python
fig = plt.figure(figsize=(3, 3))
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')
plt.show()
```

## Perform image smoothing using Gaussian filtering.

```python
# Load the image
image = cv2.imread('Week2_Q3_image.tif',0)

#display the image that is loaded
fig = plt.figure(figsize=(3, 3))
plt.imshow(image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
# Apply Gaussian filter
smoothed_image = cv2.GaussianBlur(image, (7, 7), 0) # larger the value the more pixels (7x7) will be used to smoothend, but will also increase the complexity in computation

#display the smoothened image after applying gaussian filter
fig = plt.figure(figsize=(3, 3))
plt.imshow(smoothed_image, cmap='gray')
plt.title('smoothened Image')
plt.axis('off')
plt.show()
```

## Apply median filtering to remove noise from given image.

```python
# Load the image
image = cv2.imread('Week2_Q3_image.tif',0)

#display the image that is loaded
fig = plt.figure(figsize=(3, 3))
plt.imshow(image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
# Apply median filter
median_filtered_image = cv2.medianBlur(image, 5)

#display the smoothened image after applying median filter
fig = plt.figure(figsize=(3, 3))
plt.imshow(median_filtered_image, cmap='gray')
plt.title('noise removed Image')
plt.axis('off')
plt.show()
```
