### Lab 1 - Simple Linear regression model to build b0 and b1 and show the linear model on the Graph describing the coefficient and intercept

#### include all the libraries required for this project

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

#### Load a simple data with (x,y) tuple, the X here is filtration rate and y is moisture content in a coal mining station, we are building a simple linear regression model to predict the new value and display the result

```python
# Create a NumPy array of tuples (x, y)
data_array = np.array([
    (125.3, 77.9), (98.2, 76.8), (201.4, 81.5), (147.3, 79.8), (145.9, 78.2),
    (124.7, 78.3), (112.2, 77.5), (120.2, 77.0), (161.2, 80.1), (178.9, 80.2),
    (159.5, 79.9), (145.8, 79.0), (75.1, 76.7), (151.4, 78.2), (144.2, 79.5),
    (125.0, 78.1), (198.8, 81.5), (132.5, 77.0), (159.6, 79.0), (110.7, 78.6)
])

# Printing the resulting NumPy array
#print( data_array[:, 0].reshape(-1, 1))
```

#### Split the X and y value from the dataset and pass it to LinearRegression model, we will use the model.fit to pass both the Independant and Dependant value.

#### We will use the model.predict method to calculate the mean square error, the mean square error is the average squared difference between the actual values and the predicted values made by a model. In the context of linear regression, it's commonly used to evaluate the model's performance.

![MSE Image](mse_screenshot_2.png)

```python
# Split the data into X and y sets
X = data_array[:, 0]  
y = data_array[:, 1]  

#create a simple linearRegression 
model = LinearRegression()
model.fit(X.reshape(-1, 1), y)

# Make predictions
y_pred = model.predict(X.reshape(-1, 1))

# Calculate Mean Squared Error
mse = mean_squared_error(y, y_pred)
```

#### We are plotting the data on a linear graph, the red line is the linear line which is the X prediction and the blue dots are the the x, y co-ordinates the difference beteen the red linear line and the blue dot is called the Epsilon, the bigger the difference the more the data is apart.

```python
# Plot the data points and the linear regression line
plt.scatter(X, y, label='Training Data', color='b')
plt.plot(X, model.predict(X.reshape(-1, 1)), color='r', label='Linear Regression Line')
plt.xlabel('filtration rate (X)')
plt.ylabel('moisture content (y)')
plt.legend()
plt.title('Linear Regression')
plt.show()
```

#### The below output shows the coefficients of the Linear regression, in the below

##### The mathematical formula for linear regression is  y = b0 + b1*x + e, where

##### Intercept (b0) : This is the y-intercept of the linear equation. It represents the predicted value of the dependent variable (y) when the independent variable (x) is 0.

##### Slope (b1): This is the coefficient of the independent variable (x) in the linear equation. It represents the change in the predicted value of the dependent variable (y) for a one-unit change in the independent variable (x).

##### e is the error term (also known as the residual errors), the part of y that can be explained by the regression model

```python
# Display the model's coefficients and performance metrics
print("Model Coefficients: ", model.coef_)
print("Model Intercept: ", model.intercept_)
print("Mean Squared Error (MSE): ", mse)
```

#### We will pass a new value, I am using some existing value to see if it predicts the value that is closer to what it has been trained, the predicted value is printed as output

```python
predicted_y = model.predict(np.array([[110.7]]))

print("Predicted y for new value:", predicted_y)
```
