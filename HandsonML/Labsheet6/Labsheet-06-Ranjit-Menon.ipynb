```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms.community import girvan_newman
from networkx.algorithms import community 
```

### Construct a random graph with 25 nodes and 200 edges, using GNP model, and apply Girvan-Newman algorithm to find communities

The code generates a random graph with 25 nodes and approximately 200 edges using the GNP model. It then applies the Girvan-Newman algorithm to detect communities and visualizes the graph with community colors.

```python
n_nodes = 25  #configure the nodes we would like to have
# Probability of edge creation, we need to try with this to get 200 edge, 
#but because we are using random, this will sometime go above 200 or below 200
p = 0.658  

random_graph = nx.gnp_random_graph(n_nodes, p)

# Apply Girvan-Newman algorithm to find communities
communities = girvan_newman(random_graph)

# Convert the result to a list of communities, this is to group by community
communities_list = [tuple(sorted(community)) for community in next(communities)]

print("Communities found by Girvan-Newman algorithm:")
for i, community in enumerate(communities_list, start=1):
    print(f"Community {i}: {community}")
    
#Print the node and edge details
print(f"No of Nodes : {len(random_graph.nodes)}")
print(f"No of Edges : {len(random_graph.edges)}")


pos = nx.spring_layout(random_graph)

# Visualize the graph with communities
plt.figure(figsize=(8, 8))
nx.draw_networkx_nodes(random_graph, pos, node_size=400, node_color='Green')
nx.draw_networkx_edges(random_graph, pos, alpha=0.5)
nx.draw_networkx_labels(random_graph, pos)

plt.title("Random Graph with Girvan-Newman Communities")
plt.show()
```

The code defines a graph, applies the Girvan-Newman algorithm to find communities, prints and visualizes the communities using NetworkX and Matplotlib. The graph was provided for this problem and it has to be created by setting the tuple for the Graph creation

```python
#build a tuple to generate a graph as asked us to build one from image
G2 = nx.Graph([(1, 3), (2, 3),(3,7), (6,7),(4,6),(5,6),(7,8),(8, 9),(9,11),(9,10),(8, 12),(12,13),(12,14)])

pos = nx.spring_layout(G2)

#generate the communities using girvan_newman algorithm
communities_generator = girvan_newman(G2)

#top_level_communities = next(communities_generator)
communities_list = [tuple(sorted(community)) for community in next(communities_generator)]

#sorted_communities = sorted(map(sorted, top_level_communities))
print("Communities found by Girvan-Newman algorithm:")
for i, community in enumerate(communities_list, start=1):
    print(f"Community {i}: {community}")

#visualize the graph
fig, ax = plt.subplots(figsize=(8, 8))
nx.draw(G2, ax=ax, pos=pos, node_size=1500, node_color="red",with_labels=True)
plt.show()
```
