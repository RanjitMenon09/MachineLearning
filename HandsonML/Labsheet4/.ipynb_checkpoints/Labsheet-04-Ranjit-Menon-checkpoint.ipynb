```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
import numpy as np
from sklearn.decomposition import PCA
```

#### The Abalone dataset is physical measurements taken from several thousand Tasmanian abalone. (An abalone is a sea creature much like a clam or an oyster.)

```python
abalone_df = pd.read_csv('abalone.csv')
abalone_df.sample(5)
```

```python
abalone_df.describe()
```

#### Display the dataset in a pairplot to show how this data is related to each other

#### The dataset exhibits significant correlations among its features. Specifically, there is a strong linear correlation between the length and diameter features. Additionally, the number of rings is positively correlated with various quadratic features, indicating a potential relationship between these features and the maturity of the subject.

```python
# Define a custom color palette
custom_palette = sns.color_palette("Set2", n_colors=len(abalone_df['Type'].unique()))

sns.pairplot(abalone_df, hue="Type", palette=custom_palette, height=2.5, markers=["o", "s", "D"],diag_kind="hist")
plt.suptitle("Relationship of Abalone Data", y=1.02)
plt.show()
```

#### The below correlation matrix help us to know how the data is correlated, the diagonal value shows as 1 which means it is heavily correlated and that is because it is compared to its own field, but th eone that is above to 0.80 is more related to each other, from the below we can see the Rings are not very correlated with other fields, may be because it is a target field and not one feature does not have direct relation to this field. Type is removed from this as that is a not a numeric field.

```python
# Compute the correlation matrix
correlation = abalone_df.corr(numeric_only=True)

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Create a heatmap with more customized settings
sns.heatmap(correlation, square=True, annot=True, vmin=0, vmax=1, cmap='viridis', linewidths=0.5)
plt.title('Correlation Matrix of Abalone Data')


plt.show()
```

#### We will start reducing the dimension using PCA , first step is to get hte correlation matrix, we will remove two of the fields, one is a Type field as that is not a numeric number (one solution to include this could be to convert this into 1-3 numeric value) and another is Rings which I beleive it is more of a Target (Y), so the rest of the fields will be treated as X feature

#### we will convert this into array using numpy and in this process we are normalizing this to be in the range of 0 to 1

```python
X = abalone_df.drop(['Rings','Type'], axis = 1)
X_std = StandardScaler().fit_transform(X)
X_std
corr_mat = np.corrcoef(X_std.T)
corr_mat
```

#### Eigenvalues and eigenvectors reveal covariance matrix characteristics. They help compute principal components and explain variance in the data by each variable.

```python
eigenvalues, eigenvectors = np.linalg.eig(corr_mat)
print('\nEigenvalues \n%s' %eigenvalues)
print('\nEigenvectors \n%s' %eigenvectors)
```

#### The eigen vector we get should be sorted as the highest value should be considered as PCA 1 and so on, we will be using this below to find the colums which add more value from the rest of the columns, this is how we will be reducing the dimension.

```python
# Pair eigenvalues with their corresponding eigenvectors
eigenvalue_vector_pairs = [(np.abs(eigenvalue), eigenvector) for eigenvalue, eigenvector in zip(eigenvalues, eigenvectors.T)]

# Sort the pairs based on the absolute values of eigenvalues in descending order
sorted_pairs = sorted(eigenvalue_vector_pairs, key=lambda x: x[0], reverse=True)

sorted_pairs
```

#### This code calculates the explained variance for each eigenvalue, representing the proportion of total variance explained by each principal component.

```python
tot = sum(eigenvalues)
explained_variance = [(i/tot) for i in sorted(eigenvalues, reverse=True)]
print(explained_variance)
```

#### This code computes the cumulative explained variance from a list of individual explained variance values for principal components.

```python
cumalative_var_explained = np.cumsum(explained_variance)
cumalative_var_explained
```

#### Create a bar chart to explain the variance for each principal component, providing insights into their contributions. From the below chart we can see in PCA 1 we have most of the values and most of the important features are captured till PCA 4.

```python
with plt.style.context('fivethirtyeight'):
    plt.figure(figsize=(6, 4))
    plt.bar(range(1, len(explained_variance) + 1), explained_variance, alpha=0.7, align='center', label='Explained Variance')
    plt.xlabel('Principal components')
    plt.ylabel('Explained Variance Ratio')
    plt.title('Explained Variance by Principal Components')
    plt.legend(loc='best')
    plt.tight_layout()

plt.show()
```

#### This below code compares cumulative explained variance plots for principal components, visualizing their contributions to the total variance in the dataset. We can see from PCA 4 the line is getting stable and there no improvement, so we can conclude that most of the features are created upto PCA 4.

```python
fig = plt.figure(figsize=(10, 4))

plt.plot(np.cumsum(explained_variance))

# Set x-axis limits using a single argument (min, max)
plt.xlim([0, 5])

plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative explained variance')
plt.title('Explained Variance')

plt.show()
```
